```python
import matplotlib.pyplot as plt
import pandas as pd
import os
from collections import Counter
```

```python
data=pd.read_csv(os.getcwd()+'/books_meloman.csv').to_numpy()
```

```python
genres = Counter(data[:,4].tolist())
publishers = Counter(data[:,5].tolist())
years = data[:,7].tolist()
languages = Counter(data[:,8].tolist())
ages = Counter(data[:,6].tolist())
```

```python
def plotStrings(array):
    return pd.DataFrame.from_dict(array, orient='index').plot(y=0,figsize=(15,10),kind="bar")
```

```python
plotStrings(ages)
```
